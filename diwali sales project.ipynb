```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb
```

```python
df=pd.read_csv("d:/Diwali Sales Data.csv",encoding="unicode_escape")#it gives the whole data
#to avoid encoding error using unicode_escape
```

```python
df.head(5)
```

```python
df.info()
```

```python
#to delete blank columns which is(status,unnamed1)
df.drop(['Status','unnamed1'],axis=1,inplace=True)#implace is used to save resultant data
```

```python
df.info()
```

```python
#to check missing values in col elements
pd.isnull(df)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
#drop null value-------it reduces no of rows
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
#to change datatype
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
#to change col name
df.rename(columns={'Marital_Status':'Shadi'})
```

```python
#describe method returns description of the data in the dataFrame(count,mean,std)
df.describe()
```

```python
#use describe() for specific columns
df[['Age','Orders','Amount']].describe()
```

```python
 #EDA----Exploratory data cleaning
```

```python
df.Gender.value_counts()
```

```python
x=sb.countplot(x='Gender',data=df)
for bars in x.containers:
    x.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_bar=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.barplot(x='Gender',y='Amount',data=sales_bar) 
```

Most of the purchasers are female and purchasing power of female is also more than male

```python
sb.countplot(data=df,x='Age Group',hue='Gender')
```

```python
x=sb.countplot(data=df,x='Age Group',hue='Gender')
for bars in x.containers:
    x.bar_label(bars)
```

```python
#total amount vs age group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sb.barplot(x='Age Group',y='Amount',data=sales_age)
```

most of the buyer are from group age of 26-35


STATE

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)#top ten
```

```python
sb.set(rc={'figure.figsize':(18,5)})
```

```python
sb.barplot(data=sales_state,x='State',y='Orders')
```

```python
#total amount/sales from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sb.set(rc={'figure.figsize':(15,5)})
sb.barplot(data=sales_state,x='State',y='Amount')
```

from above graph most of the order and total sales are from uttar pradesh,maharastra and karnatka respectively

#MARITAL STATUS

```python
b=sb.countplot(data=df,x='Marital_Status')
sb.set(rc={'figure.figsize':(5,5)})
for bars in b.containers:
    b.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.set(rc={'figure.figsize':(5,5)})
sb.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

Married women having purchasing power more

OCCUPATION

```python
sb.set(rc={'figure.figsize':(25,5)})
z=sb.countplot(data=df,x='Occupation')
for bars in z.containers:
    z.bar_label(bars)
```

```python
sales_occu=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.set(rc={'figure.figsize':(25,5)})
sb.barplot(data=sales_occu,x='Occupation',y='Amount')
```

from above graph purchasing power of it sector,healthcare and aviation sectors are more than others

CATEGORY

```python
sb.set(rc={'figure.figsize':(30,5)})
x=sb.countplot(data=df,x='Product_Category')
for bars in x.containers:
    x.bar_label(bars)
```

```python
sales_cat=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.set(rc={'figure.figsize':(30,5)})
sb.barplot(x='Product_Category',y='Amount',data=sales_cat)
```

from the above graph we can observe that buyer expand more money on food clothing and electronics gadgets

Conclusion:From the above data we can conclude that married women age group of 26-35 working in IT sector is likely to having more purchasing power.
